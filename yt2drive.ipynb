```python
# @title 🔍 Step 1: Check Available Formats
video_url = "https://www.youtube.com/watch?v=GYmVgweZz-k" # @param {type:"string"}
cookie_path = "/content/drive/MyDrive/JKT48/cookies.txt" # @param {type:"string"}
use_cookies = True # @param {type:"boolean"}

cookie_flag = f'--cookies "{cookie_path}"' if use_cookies else ""

# The --js-runtimes deno flag is the key fix here
!yt-dlp -F {cookie_flag} --js-runtimes deno "{video_url}"
```

```python
# @title 📥 Step 2: Download Selected IDs
video_id = "137" # @param {type:"string"}
audio_id = "140" # @param {type:"string"}
save_path = "/content/drive/MyDrive/JKT48/" # @param {type:"string"}

os.makedirs(save_path, exist_ok=True)
cookie_flag = f'--cookies "{cookie_path}"' if use_cookies else ""

print(f"🚀 Downloading Video {video_id} + Audio {audio_id} using Deno...")

!yt-dlp -f "{video_id}+{audio_id}" \
        {cookie_flag} \
        --js-runtimes deno \
        --merge-output-format mp4 \
        -o "{save_path}%(title)s.%(ext)s" \
        "{video_url}"

print("✅ Success! Check your Drive folder.")
```
