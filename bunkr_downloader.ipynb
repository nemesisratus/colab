```python
# @title Move Files
import os
import shutil

# Define your paths
source_dir = "/content/BunkrDownloader/Downloads"
target_location = "/content/"  # @param {type:"string"}

# 1. Check if the source directory exists
if not os.path.exists(source_dir):
    print(f"❌ Source directory not found: {source_dir}")
else:
    # 2. Create the target directory if it doesn't already exist
    os.makedirs(target_location, exist_ok=True)

    # 3. Move all items (files and subdirectories)
    moved_count = 0
    for item in os.listdir(source_dir):
        source_path = os.path.join(source_dir, item)
        target_path = os.path.join(target_location, item)

        try:
            shutil.move(source_path, target_path)
            print(f"✅ Moved: {item}")
            moved_count += 1
        except Exception as e:
            print(f"❌ Error moving {item}: {e}")

    print(f"\n🎉 Successfully moved {moved_count} items to {target_location}")
```
